```python

import pandas as pd

  
file =('cardiovascular-disease-dataset/\
cardio_train.csv')
```

```python
! pip install pathlib2
```

```python
# the replace code is inspired from: https://www.geeksforgeeks.org/python-program-to-replace-text-in-a-file/
import re
with open(file,'r+') as f:
  ff = f.read()
  ff = re.sub(';',',', ff)
  
        # Setting the position to the top
        # of the page to insert data
  f.seek(0)
          
        # Writing replaced data in the file
  f.write(ff)
  
        # Truncating the file size
  f.truncate()
  

```
